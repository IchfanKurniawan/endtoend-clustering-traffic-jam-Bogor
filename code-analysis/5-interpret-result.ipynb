## Group by Weekday

```python
sel_cols = ['street', 'cluster']

jams_bogor_intre_0 = jams_bogor_intre.query('weekday == 0')[sel_cols]
jams_bogor_intre_1 = jams_bogor_intre.query('weekday == 1')[sel_cols]
jams_bogor_intre_2 = jams_bogor_intre.query('weekday == 2')[sel_cols]
jams_bogor_intre_3 = jams_bogor_intre.query('weekday == 3')[sel_cols]
jams_bogor_intre_4 = jams_bogor_intre.query('weekday == 4')[sel_cols]
jams_bogor_intre_5 = jams_bogor_intre.query('weekday == 5')[sel_cols]
jams_bogor_intre_6 = jams_bogor_intre.query('weekday == 6')[sel_cols]
```

```python
def street_cluster_mode(dataframe):
    return (
                dataframe
                .groupby('street')
                .apply(lambda x: x.value_counts().head(1))
                .droplevel(0)
                .reset_index()
                .rename(columns = {0: 'occurances'})
            )
```

```python
street_cluster_weekday_0 = street_cluster_mode(jams_bogor_intre_0)
street_cluster_weekday_1 = street_cluster_mode(jams_bogor_intre_1)
street_cluster_weekday_2 = street_cluster_mode(jams_bogor_intre_2)
street_cluster_weekday_3 = street_cluster_mode(jams_bogor_intre_3)
street_cluster_weekday_4 = street_cluster_mode(jams_bogor_intre_4)
street_cluster_weekday_5 = street_cluster_mode(jams_bogor_intre_5)
street_cluster_weekday_6 = street_cluster_mode(jams_bogor_intre_6)
```

```python
street_geo = jams_prep[['street', 'geometry']].drop_duplicates()
street_geo = street_geo.sort_values('street')
```

```python
street_geo = street_geo.groupby('street').first()
```

```python
street_cluster_weekday_0 = street_cluster_weekday_0.merge(street_geo, how='left', on='street')
street_cluster_weekday_1 = street_cluster_weekday_1.merge(street_geo, how='left', on='street')
street_cluster_weekday_2 = street_cluster_weekday_2.merge(street_geo, how='left', on='street')
street_cluster_weekday_3 = street_cluster_weekday_3.merge(street_geo, how='left', on='street')
street_cluster_weekday_4 = street_cluster_weekday_4.merge(street_geo, how='left', on='street')
street_cluster_weekday_5 = street_cluster_weekday_5.merge(street_geo, how='left', on='street')
street_cluster_weekday_6 = street_cluster_weekday_6.merge(street_geo, how='left', on='street')
```
